```python
#import important modulaes
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
print("completed")
```

```python
#load data and show data head
data=pd.read_csv("vgsales.csv")
print(data.head())
```

```python
#describe data
print(data.describe())
```

```python
#check missing data
check_missing_data = data.isnull().sum()
print(check_missing_data)
```

```python
#remove null values and check it 
data['Publisher'].fillna('NaN', inplace = True)
data.dropna(inplace = True)
check_missing_data = data.isnull().sum()
print(check_missing_data)
```

```python
#what is common game type
game_type=data["Genre"].value_counts()
print (' Most common game type :- {0}' .format(game_type.index[0]))
plt.pie(game_type[0:len(game_type)],labels=game_type.index,colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']
,shadow=True,explode=[0.05]*len(game_type),autopct='%1.1f%%')
plt.title("common game type ")
plt.axis("equal")
plt.show()    
```

```python
#what is the most game type get sales
sales_sort=data.sort_values("Global_Sales")
print("most game type get sales is:- ( {0} )\n".format(sales_sort["Genre"].iloc[-1]))
y=(sales_sort['Genre'].iloc[-7:].values)
x=sales_sort["Global_Sales"].iloc[-len(y):].values
plt.barh(y=y,width=x,color = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897'])
plt.title("top five game type in sales",fontsize=14)
plt.xlabel('Global Sales',fontsize=14)
plt.ylabel('Genre',fontsize=14)
plt.show()
```

```python
#what is the most game get sales
sales_sort=data.sort_values("Global_Sales")
print("most game get sales is:- ( {0} )\n".format(sales_sort["Name"].iloc[-1]))
y=(sales_sort['Name'].iloc[-10:].values)
x=sales_sort["Global_Sales"].iloc[-10:].values
plt.barh(y=y,width=x,color = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897'])
plt.title("top 10 games in sales",fontsize=14)
plt.xlabel('Global Sales',fontsize=14)
plt.ylabel('Name',fontsize=14)
for xx,yy in zip(x,y):
  plt.text(xx,yy,xx)
plt.show()
```

```python
#what is popular platform
game_platform=data["Platform"].value_counts()
print (' Most common Platform :- {0}' .format(game_platform.index[0]))
plt.bar(game_platform.index[0:5],game_type[0:5],color = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897'])
plt.title(" the most 5 popular platform")
plt.xlabel('platform')
plt.ylabel('num of games')
plt.show()
```

```python
#what is the most game type get sales
sales_sort=data.sort_values("Global_Sales")
print("most game type get sales is:- ( {0} )\n".format(sales_sort["Platform"].iloc[-1]))
y=(sales_sort['Platform'].iloc[-16:].values)
x=sales_sort["Global_Sales"].iloc[-len(y):].values
plt.barh(y=y,width=x,color = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897'])
plt.title("top five Platform in sales",fontsize=14)
plt.xlabel('Global Sales',fontsize=14)
plt.ylabel('Platform',fontsize=14)
plt.show()
```

```python
#year_info
Year_info=data["Year"].value_counts()
print("common Year is        :-    ",round(Year_info.index[0]))
print("most_recent year is   :-    ",round(max(data["Year"])))
print("Earliest year is      :-    ",round(min(data["Year"])))
plt.pie(Year_info.values[0:5],labels=Year_info.index[0:5],colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']
,shadow=True,explode=[0.05]*5,autopct='%1.1f%%')
plt.title("popular years in number of games")
plt.axis("equal")
plt.show()
```

```python
#what is most year get sales
sales_sort=data.sort_values("Global_Sales")
print("most year get sales is:- ( {0} )\n".format(round(sales_sort["Year"].iloc[-1])))
plt.bar(sales_sort["Year"],sales_sort["Global_Sales"],color = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']
)
plt.title("years and sales")
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
#popular publishers 
publisher_info=data["Publisher"].value_counts()
print("the most popular publisher :-    ",publisher_info.index[0])
plt.pie(publisher_info.values[0:5],labels=publisher_info.index[0:5],colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']
,shadow=True,explode=[0.05]*5,autopct='%1.1f%%')
plt.title("popular publishers in number of games")
plt.axis("equal")
plt.show()
```

```python
#what is the most game get sales
sales_sort=data.sort_values("Global_Sales")
print("most publisher get sales is:- ( {0} )\n".format(sales_sort["Publisher"].iloc[-1]))
y=(sales_sort['Publisher'].iloc[-50:].values)
x=sales_sort["Global_Sales"].iloc[-50:].values
plt.barh(y=y,width=x,color = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897'])
plt.title("top 5 Publisher in sales",fontsize=14)
plt.xlabel('Global Sales',fontsize=14)
plt.ylabel('Publisher',fontsize=14)

plt.show()
```

```python
#most game in NA_Sales ,EU_Sales,JP_Sales and Other_Sales
sales_sort=data.sort_values("Global_Sales")
na_sort=data.sort_values("NA_Sales")
eu_sort=data.sort_values("EU_Sales")
jp_sort=data.sort_values("JP_Sales")
Other_sort=data.sort_values("Other_Sales")
print("most game get NA_sales is    :- ( {0} )\n".format(na_sort["Name"].iloc[-1]))
print("most game get eu_sales is    :- ( {0} )\n".format(eu_sort["Name"].iloc[-1]))
print("most game get jp_sales is    :- ( {0} )\n".format(jp_sort["Name"].iloc[-1]))
print("most game get Other_sales is :- ( {0} )\n".format(Other_sort["Name"].iloc[-1]))
y=(sales_sort['Name'].iloc[-20:].values)
x=sales_sort["NA_Sales"].iloc[-20:].values
b1=plt.barh(y=y,width=x,color = '#4F6272')
y=(sales_sort['Name'].iloc[-20:].values)
x=sales_sort["EU_Sales"].iloc[-20:].values
b2=plt.barh(y=y,width=x,color ='#B7C3F3')
y=(sales_sort['Name'].iloc[-20:].values)
x=sales_sort["JP_Sales"].iloc[-20:].values
b3=plt.barh(y=y,width=x,color ='#DD7596')
y=(sales_sort['Name'].iloc[-20:].values)
x=sales_sort["Other_Sales"].iloc[-20:].values
b4=plt.barh(y=y,width=x,color ='#8EB897')
plt.title("games & sales",fontsize=14)
plt.xlabel('NA ,EU ,JP & Other Sales',fontsize=14)
plt.ylabel('Name',fontsize=14)
plt.legend([b1,b2,b3,b4], ["NA_Sales","EU_Sales","JP_Sales","Other_Sales"])
plt.show()
```
